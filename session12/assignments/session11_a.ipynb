## Target:  draw ZIGZAG or CYCLIC TRIANGLE plot 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
cycles = 8
step_size = 50
```

```python
total_iterations = step_size * 2 * cycles
```

```python
lr_min = .001
lr_max = .010
```

```python
def get_lr(current_iter):
    """
    Returns learning rate for current iteration
    """
    cycle = np.floor(1 + current_iter / (2 * step_size))
    x = np.abs( current_iter/step_size - 2 * cycle + 1 )
    lr_t = lr_min + (lr_max - lr_min) * (1 - x)
    return lr_t
```

```python
x = np.linspace(0, total_iterations, 1000);
y = np.array([get_lr(curr_iter) for curr_iter in x])
y_min = np.array([lr_min for curr_iter in x])
y_max = np.array([lr_max for curr_iter in x])
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12,6))
plt.plot(x, y)
plt.plot(x, y_min)
plt.plot(x, y_max)
plt.xlabel('Iterations')
plt.ylabel('Learning Rate')
plt.show()
```
